```python
import numpy as np
import pandas as pd
import folium as fo
import io
%matplotlib inline
```

```python
#With the next command, Foluim will create us a complete globe (World map)
map = fo.Map() #Create complete globe
map            #Show our just created map
```

```python
#Here will create an "instance of our map" so we can add new characteristics (Childs)
x = fo.FeatureGroup(name = 'My Map')

#Next will test Geolocation by adding our child with our desired latitud and longitud
#Location : Add the place we want to mark
#Popoup   : Add a pop up message when we click on our marker
#Icon     : Icon shown in the map as marker
x.add_child(fo.Marker(location = [23, 100], popup = 'Place!', icon = fo.Icon(color = 'blue')))
```

```python
#Add to our map, our created child (Place marker)
map.add_child(x)
```

```python
#One final test!
#We will create an array of diferents Latitudes and Longitudes, then wil be shown on the map
#by itearting the array with a "for"
for lat, lon in ([34,53],[24,-50],[90,-68]):
    x.add_child(fo.Marker(location = [lat, lon], popup = 'Im here!!', icon = fo.Icon(color = 'red')))
map.add_child(x)
```

```python
#CORONAVIRUS TIME!
#Will create a new map for plot CONV-19
convMap = fo.Map()

#Next will read our dataSet and show it to see what columns will be need:
convData = pd.read_csv('ncovDeaths.csv')
convData.head()
```

```python
#As you can see, we need 4 columns minimum:
#Closest date to our current date : 2/13/20 21:15
#Lat
#Long
#Province/State
latConv = list(convData['Lat'])
lonConv = list(convData['Long'])
placeConv = list(convData['Province/State'])
countryConv = list(convData['Country/Region'])
deathsConv = list(convData['2/13/20 21:15'])
```

```python
#Now create another instance for add our new conv childs
convChild = fo.FeatureGroup(name = "Conv map")

for lat, lon, place, deaths, country in zip(latConv,lonConv, placeConv, deathsConv, countryConv):
    convChild.add_child(fo.Marker(location = [lat, lon], 
                                  popup = str(country) + " " + str(place) + " " + str(deaths), 
                                  icon = fo.Icon(color = 'red')))
```

```python
#Finally add our child to our conv map
convMap.add_child(convChild)
convMap
```
